```python
from matplotlib import pyplot as plt
import numpy as np

a = 5
b = 3
theta = np.linspace(0, 2 * np.pi, 100)
x = a * np.cos(theta)
y = b * np.sin(theta)
plt.plot(x, y)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
```

```python
from matplotlib import pyplot as plt
import numpy as np

a = 1
b = 1
c = 2
x = np.linspace(-4, 3, 100)
y = (a *x**2)+(b*x)+c

plt.plot(x, y)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
```
